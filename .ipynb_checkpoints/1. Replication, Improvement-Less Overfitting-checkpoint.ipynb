# Replicate previous work

Run `GridSearchCV` to optimize `n_estimators` for Random Forest classifier

````python
rf = RandomForestClassifier()
rf_params = {'n_estimators': [10, 20, 50, 100, 200, 400]}
gsRf = GridSearchCV(estimator=rf, param_grid=rf_params)
````

````python
gsRf.fit(X_train, y_train)
gsRf.best_params_ 
````

Output: {'n_estimators': 400}

````python
gsRf.predict(X_test)
gsRf.score(X_test, y_test) # Accuracy
````

Output: 0.6662182613082226

````python
gsRf.score(X_train, y_train)
````

Output: 1.0
This means that the model is overfitting on the training set.

Create moving window of 10000 samples and predict on the following 1000 samples. Save actual and predictions in DataFrame and calculate accuracy

````python
index_min = 0
index_max = 10000
rf = RandomForestClassifier(n_estimators=400)
rf_predictions = pd.DataFrame(columns=['y_actual', 'y_predict'], dtype=int)
acc_training = []

while index_max <= 39000:
    X_train_rf = X.iloc[index_min : index_max]
    y_train_rf = y.iloc[index_min : index_max]
    X_test_rf = X.iloc[index_max : (index_max + 1000)]
    y_test_rf = y.iloc[index_max : (index_max + 1000)]
    rf.fit(X_train_rf, y_train_rf)
    acc_training.append(rf.score(X_train_rf, y_train_rf))
    y_predictions_rf = rf.predict(X_test_rf)
    y_actual_and_predict_rf = pd.DataFrame({'y_actual': y_test_rf, 'y_predict': y_predictions_rf})
    rf_predictions = rf_predictions.append(y_actual_and_predict_rf, ignore_index=False)
    index_min += 1000
    index_max += 1000
````

````python
from sklearn.metrics import accuracy_score
print(acc_training)
accuracy_score(rf_predictions['y_actual'].values, rf_predictions['y_predict'].values)
````

Output: 0.671704223451626

# Improve upon previous work

Use `GridSearchCV` to optimize upon more features

````python
rf_improve = RandomForestClassifier()
rf_params_improve = {'n_estimators': [200, 400, 750, 1000], 'max_depth' : [15, 20, 30], 'min_samples_leaf' : [1, 5, 10]}
gsRf_improve = GridSearchCV(estimator=rf_improve, param_grid=rf_params_improve)
````

````python
gsRf_improve.fit(X_train, y_train)
print(gsRf_improve.best_params_ )
gsRf_improve.predict(X_test)
print(gsRf_improve.score(X_test, y_test)) # Accuracy
````

Output: {'max_depth': 30, 'min_samples_leaf': 5, 'n_estimators': 750}
0.6652934252564319

````python
index_min = 0
index_max = 10000
rf = RandomForestClassifier(n_estimators=750, max_depth=30, min_samples_leaf=5)
rf_predictions_improve = pd.DataFrame(columns=['y_actual', 'y_predict'], dtype=int)
acc_training = []

while index_max <= 39000:
    X_train_rf = X.iloc[index_min : index_max]
    y_train_rf = y.iloc[index_min : index_max]
    X_test_rf = X.iloc[index_max : (index_max + 1000)]
    y_test_rf = y.iloc[index_max : (index_max + 1000)]
    rf.fit(X_train_rf, y_train_rf)
    acc_training.append(rf.score(X_train_rf, y_train_rf))
    y_predictions_rf = rf.predict(X_test_rf)
    y_actual_and_predict_rf = pd.DataFrame({'y_actual': y_test_rf, 'y_predict': y_predictions_rf})
    rf_predictions_improve = rf_predictions_improve.append(y_actual_and_predict_rf, ignore_index=False)
    index_min += 1000
    index_max += 1000
````

````python
print(acc_training)
accuracy_score(rf_predictions_improve['y_actual'].values, rf_predictions_improve['y_predict'].values)
````

Output: 0.671636756173256
